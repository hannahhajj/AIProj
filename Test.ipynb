# Environment check


```python
import sys
import importlib

print("Python executable:", sys.executable)

packages = [
    ("NumPy", "numpy"),
    ("Pandas", "pandas"),
    ("Matplotlib", "matplotlib"),
    ("Seaborn", "seaborn"),
    ("scikit-learn", "sklearn"),
    ("Plotly", "plotly"),
    ("SciPy", "scipy"),
    ("Joblib", "joblib"),
    ("ipywidgets", "ipywidgets"),
    ("RDKit", "rdkit"),
    ("Mordred", "mordred"),
    ("PubChemPy", "pubchempy"),
    ("SELFIES", "selfies"),
    ("XGBoost", "xgboost"),
    ("LightGBM", "lightgbm"),
    ("CatBoost", "catboost"),
    ("UMAP", "umap"),
    ("SHAP", "shap"),
]

for label, module_name in packages:
    try:
        module = importlib.import_module(module_name)
        version = getattr(module, "__version__", "installed (version not exposed)")
        print(f"{label}: {version}")
    except Exception as e:
        print(f"{label}: import failed -> {e}")
```
