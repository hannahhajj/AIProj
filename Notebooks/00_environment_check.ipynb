# 00 · Environment Check

Verify that all required packages are installed and importable before running any analysis notebook.

```python
import sys
import importlib

print(f"Python {sys.version}\n")

# (display name, import name, minimum version from project plan)
REQUIRED = [
    ("pandas",       "pandas",      "2.2.3"),
    ("numpy",        "numpy",       "2.2.1"),
    ("scipy",        "scipy",       "1.14.1"),
    ("statsmodels",  "statsmodels", "0.14.4"),
    ("scikit-learn", "sklearn",     "1.6.0"),
    ("matplotlib",   "matplotlib",  "3.10.0"),
    ("seaborn",      "seaborn",     "0.13.2"),
    ("bokeh",        "bokeh",       "3.9.0"),
    ("umap-learn",   "umap",        None),
    ("shap",         "shap",        None),
    ("glycowork",    "glycowork",   "1.7.1"),
    ("glycorender",  "glycorender", "0.2.5"),
    ("IPython",      "IPython",     "8.39.0"),
]

OPTIONAL = [
    ("xgboost", "xgboost", None),
]

def check_packages(pkg_list, label):
    failures = []
    print(f"── {label} ──────────────────────────")
    for name, mod, min_ver in pkg_list:
        try:
            m = importlib.import_module(mod)
            ver = getattr(m, "__version__", "(version unknown)")
            print(f"  ✓  {name:<16} {ver}")
        except ImportError as e:
            print(f"  ✗  {name:<16} MISSING — {e}")
            failures.append(name)
    return failures

required_failures = check_packages(REQUIRED, "Required")
print()
check_packages(OPTIONAL, "Optional")
```

```python
from pathlib import Path

ROOT = Path("..").resolve()

DATA_FILES = [
    ("data matrix",        "files/data/input/internship_data_matrix.csv"),
    ("feature metadata",   "files/data/input/internship_feature_metadata.csv"),
    ("acquisition list",   "files/data/input/intership_acquisition_list.csv"),
    ("exogenous standards","files/data/input/exogenous_standards.csv"),
    ("glycan list",        "files/data/glycan_embedding/glycan_list.csv"),
    ("df_glycan",          "files/data/glycan_embedding/df_glycan.pkl"),
    ("glycan_binding",     "files/data/glycan_embedding/glycan_binding.pkl"),
    ("N_glycans_df",       "files/data/glycan_embedding/N_glycans_df.pkl"),
    ("df_species",         "files/data/glycan_embedding/df_species.pkl"),
]

print("── Data files ───────────────────────────")
file_failures = []
for label, rel_path in DATA_FILES:
    p = ROOT / rel_path
    if p.exists():
        size_mb = p.stat().st_size / 1_048_576
        print(f"  ✓  {label:<22} {size_mb:>6.1f} MB   {p.name}")
    else:
        print(f"  ✗  {label:<22} NOT FOUND — expected at {p}")
        file_failures.append(label)
```

```python
print("═" * 44)
if not required_failures and not file_failures:
    print("  All checks passed. Ready to start.")
else:
    if required_failures:
        print(f"  Missing packages ({len(required_failures)}): {', '.join(required_failures)}")
        print("  → Run: pip install -r requirements.txt")
    if file_failures:
        print(f"  Missing data files ({len(file_failures)}): {', '.join(file_failures)}")
        print("  → Check that files/data/ is populated correctly")
print("═" * 44)
```
