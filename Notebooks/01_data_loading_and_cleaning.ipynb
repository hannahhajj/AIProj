```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Data Matrix (Peak Areas) - Rows:  Bio Samples (blood, biofluids), Cols: Features (chroma peaks that give intensity or abundance of that specific molecule)
data_matrix = pd.read_csv('/Users/nadimkhalife/Desktop/ml-internship/data/input/internship_data_matrix.csv', index_col=0)

# 2. Feature Metadata (Retention Time, m/z) - Rows: Features IDs (matching columns of data matrix), Cols: Descriptors
feature_metadata = pd.read_csv('/Users/nadimkhalife/Desktop/ml-internship/data/input/internship_feature_metadata.csv', index_col=0)

# 3. Acquisition List (Sample metadata, run order, batch) - Rows: Samples
acquisition_list = pd.read_csv('/Users/nadimkhalife/Desktop/ml-internship/data/input/intership_acquisition_list.csv', index_col=0)

# 4. Exogenous Standards 
standards = pd.read_csv('/Users/nadimkhalife/Desktop/ml-internship/data/input/exogenous_standards.csv', index_col=0)
```

```python
# Check the batches available
print(acquisition_list['batch'].unique())

# Filter acquisition list to ONLY include batch 1 (using the integer 1)
acq_batch1 = acquisition_list[acquisition_list['batch'] == 1].copy()

# Filter the data matrix to only keep the samples that exist in batch 1
samples_in_batch1 = acq_batch1.index
data_matrix_b1 = data_matrix.loc[samples_in_batch1].copy()

print(f"Original samples: {len(data_matrix)}")
print(f"Batch 1 samples: {len(data_matrix_b1)}")

# Some Sanity checks 

# The number of samples (rows) in data_matrix should exactly match acquisition_list
assert data_matrix_b1.shape[0] == acq_batch1.shape[0]

# The number of features (columns) in data_matrix should exactly match rows in feature_metadata
assert data_matrix_b1.shape[1] == feature_metadata.shape[0]

print("Data alignment verified. Ready for EDA!")
```
